# YOLOv8 Out-of-stock and Price Tag Detectors


---





```python
def oos_detector_v8():
    """
    Perform out-of-stock detection using a custom YOLOv8 model.

    Returns:
        DataFrame: A DataFrame containing the detected out-of-stock coordinates.
    """

  # load yolov8 custom model
  model = YOLO('/content/oos_weights_v8.pt')

  # create empty dataframe to store results
  colnames = ['xmin', 'ymin', 'xmax', 'ymax']
  df = pd.DataFrame(columns=colnames + ['image'])

  # iterate through all images
  for imgName in os.listdir('/content/test_imgs/test'):
    # make prediction
    results = model.predict('/content/test_imgs/test/' + imgName, save=True, line_width=15)

    # iterate through all resulting bounding boxes
    for r in results:
      # check if result is non-null
      if len(r) > 0:
        # extract bounding box coordinates and store in dataframe
        results_df = pd.DataFrame(r.boxes.cpu().xyxy.numpy(), columns = colnames)
        # store corresponding image name in image column
        results_df['image'] = imgName
        df = df.append(results_df)

  return df
```

```python
def pt_detector_v8():
      """
    Perform price tag detection using a custom YOLOv8 model.

    Returns:
        DataFrame: A DataFrame containing the detected price tag coordinates.
    """

  # load yolov8 custom model
  model = YOLO('/content/pt_weights_v8.pt')
  # create empty dataframe to store results
  colnames = ['xmin', 'ymin', 'xmax', 'ymax']
  df = pd.DataFrame(columns=colnames + ['image'])

  # iterate through all images
  for imgName in os.listdir('/content/test_imgs/test'):
    # make prediction
    results = model.predict('/content/test_imgs/test/' + imgName, save=True, line_width=15)

    # iterate through all resulting bounding boxes
    for r in results:
      # check if result is non-null
      if len(r) > 0:
        # extract bounding box coordinates and store in dataframe
        results_df = pd.DataFrame(r.boxes.cpu().xyxy.numpy(), columns = colnames)
         # store corresponding image name in image column
        results_df['image'] = imgName
        df = df.append(results_df)

  return df
```

# Matching Out-of-stock to Price Tags

---

```python
def find_closest_tag(xb, yb, oos_ymin, tags):
    '''(float, float, float, pandas.dataframe) -> list, int
    Receives bottom center (xb,yb) and ymin coordinates from an OOS bounding box,
    as well as a dataset containing columns for the image name and as well as
    (xmin, ymin, xmax, and ymax) coordinates.
    Returns coordintes of the closest price tag to the OOS bbox in the image
    and the corresponding distance.
    '''

    # get the top center coordinates from the first tag in an image
    closest_tag = [tags['xmin'].values[0] , tags['ymin'].values[0], tags['xmax'].values[0], tags['ymax'].values[0]]
    # get the distance between OOS and first tag
    min_dist = math.dist((xb,yb), ((closest_tag[0]+closest_tag[2])/2, closest_tag[1]))

    # iterate through all price tags to find the one with minimum distance
    for t in range(len(tags)):
        # get tag coordinates
        xmin, ymin, xmax, ymax = tags['xmin'].values[t], tags['ymin'].values[t], tags['xmax'].values[t], tags['ymax'].values[t]

        # check if tag is below OOS
        if ymin >= oos_ymin:
            # get the distance between bottom center oos and top center tag points
            euc_dist = math.dist((xb,yb), ((xmin+xmax)/2, ymin))
            # if the distance is smaller than current minimum, upate min_dist
            if euc_dist < min_dist:
              closest_tag = [xmin, ymin, xmax, ymax]
              min_dist = euc_dist

    # if no tag below oos was found, return nan for coordinates and distance
    if closest_tag[1] <= oos_ymin:
      return [np.nan]*4, np.nan

    return closest_tag, min_dist
```

```python
def oos_tag_match(df_oos, df_tag):
    '''(pandas.dataframe, pandas.dataframe) -> pandas.dataframe

    Receives 2 dataframe as input:
    - df_oos:  contains information about out-of-stock bounding boxes in an image.
    It should have columns for the image name and the coordinates of the bottom-left and top-right
    corners of the OOS bounding boxes (xmin, ymin, xmax, and ymax).
    - df_tag:  contains information about price tag bounding boxes in all images.
    It should have columns for the image name and as well as (xmin, ymin, xmax, and ymax).

    Returns a new DataFrame (df_match) with additional columns for the matched price tag bounding boxes'
    coordinates (xmin_tag, ymin_tag, xmax_tag, and ymax_tag) for each corresponding OOS bounding box.
    '''

    # create a copy of df_oos to store the matched results
    df_match = df_oos.copy()

    # initialize new columns in df_match with NaN values to store the matched price tag coordinates
    df_match['xmin_tag'] = np.nan
    df_match['ymin_tag'] = np.nan
    df_match['xmax_tag'] = np.nan
    df_match['ymax_tag'] = np.nan

    # iterate through each row of df_oos
    for i in range(len(df_oos)):
        # for each OOS bounding box, extract the image name
        img = df_oos['image'][i]
        # find all the price tag bounding boxes (img_tags) in df_tag that belong to the same image
        img_tags = df_tag[df_tag['image'] == img]


        if len(img_tags) != 0:
            # if there are any matching price tags, get bottom center coordinates (xb, yb) of the OOS bounding box
            xb, yb = (df_oos['xmin'].values[i] + df_oos['xmax'].values[i])/2, df_oos['ymax'].values[i]
            # find closest price tag and update df_match with the matched price tag coordinates
            closest_tag, dist = find_closest_tag(xb, yb, df_oos['ymin'].values[i], img_tags)
            df_match.loc[i, 'xmin_tag'] = closest_tag[0]
            df_match.loc[i, 'ymin_tag'] = closest_tag[1]
            df_match.loc[i, 'xmax_tag'] = closest_tag[2]
            df_match.loc[i, 'ymax_tag'] = closest_tag[3]

    return df_match
```

```python
def draw_match(df):
    """
    Draw bounding boxes on images to visualize matches between out-of-stock and price tag coordinates.

    Args:
        df (DataFrame): The DataFrame containing match information.

    Returns:
        None
    """
  # Iterate through all images in the test directory
  for imgName in os.listdir('/content/test_imgs/test'):
    # Filter DataFrame to get matching rows for the current image
    df_img = df[df['image'] == imgName]
    # Read the image
    img = read_image('/content/test_imgs/test/' + imgName)

    bbs = []

    num_oos = len(df_img)
    # Check if image contain OOS
    if num_oos == 0:
      print(f"No OOS detected in image {imgName}.")
    else:
      # Loop through out-of-stock items and their corresponding price tag coordinates
      for i in range(num_oos):
        # Append out-of-stock bounding box coordinates
        bbs.append([df_img['xmin'].values[i], df_img['ymin'].values[i], df_img['xmax'].values[i], df_img['ymax'].values[i]])
         # Append price tag coordinates if available
        tag_coords = [df_img['xmin_tag'].values[i], df_img['ymin_tag'].values[i], df_img['xmax_tag'].values[i], df_img['ymax_tag'].values[i]]
        if tag_coords[0] == np.nan:
          print(f"No price tags detected in image {imgName}.")
        else:
          bbs.append(tag_coords)

      # Convert bounding box coordinates to a torch tensor
      bbox = torch.tensor(bbs, dtype=torch.float)
      # Draw bounding boxes on the image
      img_w_bbox = draw_bounding_boxes(img, bbox, width=15, colors=(255,0,0))
      # Convert the image with bounding boxes to a PIL image
      img_w_bbox_pil = torchvision.transforms.ToPILImage()(img_w_bbox)
      # Save the image with drawn bounding boxes
      img_w_bbox_pil.save("/content/matched_imgs/"+ imgName)
```

# Extract Information from Price Tags

---

```python
def oos_tag_extract(matched_coords):
    """
    This function extracts information from each row of the given DataFrame and adds it to new columns.

    Args:
        matched_coords (DataFrame): The DataFrame containing bounding box coordinate of
        OOS and its corresponding price tag.

    Returns:
        None
    """

  # iterate through all rows in matched_coords dataframe
  for index, row in matched_coords.iterrows():
    # extract barcode and text in image
    price_tag_barcode, price_tag_text = process_dataframe_row(row)
    # Update the 'price_tag_barcode' column for the current row
    matched_coords.at[index, 'price_tag_barcode'] = price_tag_barcode
    # Update the 'price_tag_text' column for the current row
    matched_coords.at[index, 'price_tag_text'] = [price_tag_text]


  return matched_coords

def decode(image):
   """
    Decode barcodes or QR codes from an image using the pyzbar library.

    Args:
        image: The image data to be decoded.

    Returns:
        str: The decoded data from the barcode or QR code, or None if no code is found.
    """
    # Use the pyzbar library to decode objects (barcodes or QR codes) from the image
  decoded_objects = pyzbar.decode(image)
  # Iterate through the decoded objects
  for obj in decoded_objects:
    # Return the data of the first decoded object
    return obj.data

def text_extractor(image_path):
  """
    Extract text from an image using the EasyOCR library.

    Args:
        image_path (str): The path to the image file.

    Returns:
        list: A list of extracted text and coordinates from the image.
    """

  image = cv2.imread(image_path)
  reader = easyocr.Reader(['en'])
  result = reader.readtext(image)
  return result

def process_dataframe_row(row):
      """
    Process a row of the DataFrame to extract barcode and text information.

    Args:
        row (Series): A row of the DataFrame containing image and coordinate information.

    Returns:
        tuple: A tuple containing extracted barcode and text information.
    """

  # Extract necessary coordinates and image path from the row
  image_path = '/content/test_imgs/test/' + row['image']
  xmin, ymin, xmax, ymax = row['xmin'], row['ymin'], row['xmax'], row['ymax']
  xmin_tag, ymin_tag, xmax_tag, ymax_tag = row['xmin_tag'], row['ymin_tag'], row['xmax_tag'], row['ymax_tag']

  # Check if price tag coordinates are valid
  if math.isnan(xmin_tag) or math.isnan(ymin_tag) or math.isnan(xmax_tag) or math.isnan(ymax_tag):
    return None, None  # Return None for price_tag_barcode and price_tag_text

  # Read the image and create a masked image of only the price tag
  image = cv2.imread(image_path)
  if image is None:
    print(f"Error: Could not read the image from path: {image_path}")
    return None, None  # Return None for price_tag_barcode and price_tag_text

  #Mask everything outside the price tag coordinates
  masked_image = image[int(ymin_tag):int(ymax_tag), int(xmin_tag):int(xmax_tag)]

  #Decode the barcode and extract price tag information
  price_tag_barcode = decode(masked_image)
  price_tag_text = text_extractor(image_path)

  return price_tag_barcode, price_tag_text

```

## Send Barcode to API

Function was not added to main since the API doesn't have all the product names.

```python
import requests

def barcode_to_api(barcode_value):
      """
    Retrieve product information from the Open Food Facts API based on a barcode value.

    Args:
        barcode_value (str): The barcode value for the product.

    Returns:
        str: The product name retrieved from the API.
    """

  # Construct the URL for the API request
  url = f'https://world.openfoodfacts.org/api/v0/product/{barcode_value}.json'
  # Make a GET request to the API
  response = requests.get(url)
    # Extract the product name from the API response JSON
    try:
        product_name = response.json()['product']['product_name']
        print(f'The product name for barcode {barcode_value} is: {product_name}')
        return product_name
    except KeyError:
        print(f'Product information not found for barcode {barcode_value}')
        return None
```

# main

```python
def main():

    # Step 1: Download test images and weights
    getTestImages()
    getOOSWeights_v8()
    getPTWeights_v8()

    # Step 2: Perform out-of-stock detection and store results in 'oos_coords'
    oos_coords = oos_detector_v8().reset_index()

    # Step 3: Perform price tag detection and store results in 'pt_coords'
    pt_coords = pt_detector_v8().reset_index()

    # Step 4: Match out-of-stock coordinates with corresponding price tag coordinates
    matched_coords = oos_tag_match(oos_coords, pt_coords)

    # Step 5: Create a directory to store images with drawn matches
    os.mkdir('matched_imgs')

    # Step 6: Draw bounding boxes to visualize matches and save the images
    draw_match(matched_coords)

    # Step 7: Extract price tag information of the out-of-stock items
    extracted = oos_tag_extract(matched_coords)

    # Step 8: Save the resulting DataFrame to a compressed CSV file
    compression_opts = dict(method='zip', archive_name='extracted_df.csv')
    extracted.to_csv('extracted_df.zip', index=False, compression=compression_opts)
```

```python
if __name__ == "__main__":
  main()
```
